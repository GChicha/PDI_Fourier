```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
lena = plt.imread("lena_std.tif")[:, :, 1]
```

```python
ff_lena = np.fft.fftshift(np.fft.fft2(lena))
```

```python
plt.figure(figsize=(10, 6))

plt.imshow(np.log10(abs(ff_lena) ** 2))
```

```python
fft_messed = ff_lena.copy()

fft_messed[110, 110] = ff_lena.max() * 2
fft_messed[110, 400] = ff_lena.max() * 2
```

```python
plt.figure(figsize=(10, 6))

plt.imshow(np.log10(abs(fft_messed) ** 2))
```

```python
plt.figure(figsize=(10, 6))

plt.imshow(abs(np.fft.ifft2(fft_messed)))
```
