```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
lena = plt.imread("lena_std.tif")[:, :, 1]
```

```python
ff_lena = np.fft.fftshift(np.fft.fft2(lena))
```

```python
plt.figure(figsize=(10, 6))

plt.imshow(np.log10(abs(ff_lena) ** 2))
```

```python
def pol2cart(raio, angulo):
    x = int(((lena.shape[0] // 2) * raio * np.cos(math.radians(angulo))) + lena.shape[0] // 2)
    y = int(((lena.shape[0] // 2) * raio * np.sin(math.radians(angulo))) + lena.shape[0] // 2)
    return(x, y)
```

```python
raio = .09
angulo1 = 90
angulo2 = 80

vlr = ff_lena.max() * .4

fft_messed = ff_lena.copy()
# fft_messed = np.zeros(lena.shape)

fft_messed[pol2cart(raio, angulo1)] = vlr
fft_messed[pol2cart(raio, angulo1 + 180)] = vlr

fft_messed[pol2cart(raio, angulo2)] = vlr
fft_messed[pol2cart(raio, angulo2 + 180)] = vlr
```

```python
plt.figure(figsize=(10, 6))

plt.imshow(np.log10(abs(fft_messed) ** 2))
```

```python
plt.figure(figsize=(10, 6))

plt.imshow(abs(np.fft.ifft2(fft_messed)), cmap="gray")
```
